# What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions about a population based on sample data. It involves formulating two competing hypotheses: the null hypothesis (denoted as (H_0)) and the alternative hypothesis (denoted as (H_1) or (H_a)). The null hypothesis typically represents a statement of no effect or no difference, while the alternative hypothesis represents the statement that there is an effect or a difference.

# What is the null hypothesis, and how does it differ from the alternative hypothesis?
The null hypothesis ((H_0)) posits that there is no effect or difference, while the alternative hypothesis ((H_1)) posits that there is an effect or difference.
The null hypothesis serves as the default assumption that researchers test against, while the alternative hypothesis represents the outcome that researchers are interested in demonstrating through their data.





# What is the significance level in hypothesis testing, and why is it important?
The significance level, denoted as alpha(sign), is a threshold that helps you decide whether to reject the null hypothesis ((H_0)) in a hypothesis test.
It represents the probability of making a Type I error, which occurs when you reject the null hypothesis when it is actually true

# What does a P-value represent in hypothesis testing?
The p-value helps you decide whether to reject the null hypothesis based on your data.

A small p-value (≤ 0.05) suggests strong evidence against (H_0), leading you to reject it.

A large p-value (> 0.05) suggests insufficient evidence to reject (H_0), meaning you do not reject it.

# How do you interpret the P-value in hypothesis testing?
Small P-value (≤ (alpha), e.g., 0.05):

Indicates strong evidence against (H_0).
Action: Reject the null hypothesis. Conclude that there is a statistically significant effect.

Large P-value (> (\alpha)):

Indicates insufficient evidence to reject (H_0).
Action: Fail to reject the null hypothesis. Conclude that there is not enough evidence for an effect.

# What are Type 1 and Type 2 errors in hypothesis testing?
**Type I Error (False Positive)**

Definition: Rejecting the null hypothesis ((H_0)) when it is actually true.
Probability: Denoted by (alpha) (e.g., (alpha = 0.05) means a 5% chance).
Example: Concluding a drug is effective when it is not.

**Type II Error (False Negative)**
Definition: Failing to reject the null hypothesis when it is actually false.
Probability: Denoted by (beta).
Example: Concluding a drug is not effective when it actually is.

# What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
A one-tailed test checks for a difference in only one direction (either greater or less than), with the rejection region on one side of the distribution. It is used when we expect a specific directional change.

A two-tailed test checks for a difference in both directions (greater or lesser), with the rejection region split between both tails. It is used when we expect any significant difference, regardless of direction.

# What is the Z-test, and when is it used in hypothesis testing?
A Z-test is a method used in hypothesis testing to check if there is a significant difference between:

A sample mean and a population mean (One-Sample
Z-Test).

Two sample means (Two-Sample Z-Test).

Sample proportions (Proportion Z-Test).

It helps us decide whether a claim about a population is true or not, based on a sample.

When to Use a Z-Test?

✅ The sample size is large (n ≥ 30).
✅ The data follows a normal distribution.
✅ The population standard deviation (σ) is known.




# How do you calculate the Z-score, and what does it represent in hypothesis testing?

### **Formula for Z-Score:**  
For a **single data point**:  
\[
Z = \frac{{X -μ }}{σ}
\]
For a **sample mean (Z-test formula)**:  
\[
Z = \frac{{=X̄ - μ}}{{σ / sqrt{n}}}
\]
Where:  
- **Z** = Z-score  
- **X̄** = Sample mean  
- **μ** = Population mean  
- **σ** = Population standard deviation  
- **n** = Sample size  



### **What Does the Z-Score Represent?**  
- A **Z-score of 0** means the sample mean is **exactly equal** to the population mean.  
- A **positive Z-score** means the sample mean is **above** the population mean.  
- A **negative Z-score** means the sample mean is **below** the population mean.  
- A **large Z-score** (positive or negative) means the sample mean is **far from** the population mean, suggesting a significant difference.  

In hypothesis testing, we compare the **Z-score** to a **critical value** (from the Z-table) to decide whether to reject the **null hypothesis (H₀)**.

# What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution (also called Student’s t-distribution) is a probability distribution used in statistics when:

*   The sample size is small (n < 30).
*   The population standard deviation (σ) is unknown.


It is similar to the normal distribution but has fatter tails, meaning it accounts for more variability when the sample size is small.

Use T-distribution when:
Sample size is small (n < 30).
Population standard deviation (σ) is unknown.

Use Normal distribution when:
Sample size is large (n ≥ 30).
Population standard deviation (σ) is known.



# What is the difference between a Z-test and a T-test?
Difference Between Z-Test and T-Test
Sample Size:

Z-Test: Used for large samples (n ≥ 30).
T-Test: Used for small samples (n < 30).
Standard Deviation (σ):

Z-Test: Known population standard deviation.
T-Test: Unknown population standard deviation (use sample standard deviation ‘s’).
Distribution:

Z-Test: Follows a normal distribution.
T-Test: Follows a T-distribution (fatter tails for small samples).
Use Case:

Z-Test: Compare sample mean with population mean when σ is known.
T-Test: Compare sample mean with population mean when σ is unknown.




# What is the T-test, and how is it used in hypothesis testing?
A T-Test is a statistical test used to compare the mean of a sample to a population mean or compare the means of two groups when the population standard deviation (σ) is unknown and the sample size is small (n < 30).

How is the T-Test Used in Hypothesis Testing?
Define the Hypotheses

Null Hypothesis (H₀): There is no significant difference between the means.
Alternative Hypothesis (H₁): There is a significant difference between the means.
Choose the Type of T-Test

One-Sample T-Test: Compares a sample mean with a population mean.
Independent (Two-Sample) T-Test: Compares the means of two independent groups.
Paired T-Test: Compares the means of the same group before and after a change.
Calculate the T-Statistic
Formula: T = (X̄ - μ) / (s / √n)
Where:

X̄ = Sample Mean
μ = Population Mean
s = Sample Standard Deviation
n = Sample Size
Compare with Critical Value or p-value

If |T| > Critical Value or p-value < 0.05, reject H₀ (significant difference).
Otherwise, fail to reject H₀ (no significant difference).

# What is the relationship between Z-test and T-test in hypothesis testing?
Relationship Between Z-Test and T-Test in Hypothesis Testing
Both Z-Test and T-Test are used to compare means in hypothesis testing.

Z-Test is used when the sample size is large (n ≥ 30) and the population standard deviation (σ) is known.

T-Test is used when the sample size is small (n < 30) and the population standard deviation (σ) is unknown (sample standard deviation ‘s’ is used instead).

As sample size increases, the T-distribution approaches the normal (Z) distribution, making both tests similar for large samples.

# What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval (CI) is a range of values that estimates an unknown population parameter (like the mean) with a certain level of confidence (e.g., 95%). It gives an idea of how much uncertainty exists in the estimate.

How is a Confidence Interval Used?
Understanding Uncertainty

A 95% confidence interval means that if we repeat the study many times, 95% of the intervals will contain the true population value.

Decision Making in Hypothesis Testing

If the CI includes the null hypothesis value, we fail to reject the null hypothesis.

If the CI does not include the null hypothesis value, we reject the null hypothesis.

Comparing Two Groups

If the confidence interval for the difference between two means includes zero, there is no significant difference between the groups.

# What is the margin of error, and how does it affect the confidence interval?
The Margin of Error (MOE) is the amount of uncertainty in a confidence interval. It tells us how much the sample estimate (like the mean) could vary from the true population value.

How Does MOE Affect the Confidence Interval?

Larger MOE → Wider Confidence Interval
More uncertainty, less precision

Happens with smaller sample sizes or higher confidence levels

Smaller MOE → Narrower Confidence Interval

Less uncertainty, more precision

Happens with larger sample sizes or lower confidence levels

# What is the Chi-square distribution, and when is it used?
The Chi-Square distribution is a probability distribution used for categorical data. It is skewed and depends on degrees of freedom (df).

When is it Used?

Goodness-of-Fit Test: Checks if observed data fits an expected distribution.

# What is the Chi-square goodness of fit test, and how is it applied?
### **What is the Chi-Square Goodness of Fit Test?**  
The **Chi-Square Goodness of Fit Test** checks if **observed categorical data** matches an **expected distribution**. It helps determine if a dataset follows a known distribution.  

### **Formula:**  
\[
chi^2 = sum \frac{(O - E)^2}{E}
\]  
Where:  
- **O** = Observed frequency  
- **E** = Expected frequency  

### **How is it Applied?**  
1. **State Hypotheses:**  
   - **H₀ (Null Hypothesis):** Data follows the expected distribution.  
   - **H₁ (Alternative Hypothesis):** Data does not follow the expected distribution.  

2. **Calculate Chi-Square Value:** Compare observed and expected frequencies using the formula.  

3. **Determine Critical Value/P-Value:** Use a Chi-Square table with **degrees of freedom (df = categories - 1)**.  

4. **Make a Decision:**  
   - If **χ² > critical value** or **p-value < significance level (α)** → Reject **H₀**.  
   - Otherwise, fail to reject **H₀** (data fits the expected distribution).  


# What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is a right-skewed probability distribution used to compare variances between groups. It depends on two degrees of freedom (df₁, df₂).

When is it Used?

ANOVA (Analysis of Variance): Compares means of multiple groups by analyzing variance.

F-Test for Variance: Checks if two populations have equal variances.

Regression Analysis: Tests overall significance of regression models.

# What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) is a statistical test used to compare means of three or more groups to check if they are significantly different. It uses the F-distribution to test variance differences.

Assumptions of ANOVA:

Independence: Samples must be independent of each other.

Normality: Data in each group should be normally distributed.

Homogeneity of Variance: All groups should have equal variances (checked using an F-test).

# What are the different types of ANOVA tests?
Types of ANOVA Tests:

One-Way ANOVA → Compares means of one independent variable (factor) across multiple groups.


Two-Way ANOVA → Compares means based on two independent variables and checks for interaction effects.


Repeated Measures ANOVA → Used when the same subjects are tested under different conditions over time.

MANOVA (Multivariate ANOVA) → Analyzes multiple dependent variables simultaneously.



# What is the F-test, and how does it relate to hypothesis testing?
### **What is the F-Test?**  
The **F-test** is a statistical test used to compare **variances** between two or more groups. It follows the **F-distribution** and is commonly used in ANOVA and regression analysis.  

### **How It Relates to Hypothesis Testing:**  
1. **Null Hypothesis (H₀):** Variances (or means) are equal.  
2. **Alternative Hypothesis (H₁):** Variances (or means) are different.  
3. **Calculate F-Value:**  
   \[
   F = \frac{\text{Variance of Group 1}}{\text{Variance of Group 2}}
   \]  
4. **Compare with Critical Value/P-Value:** If **F > critical value** or **p-value < α**, reject **H₀**.

```python
#  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
# interpret the results

import scipy.stats as stats
import numpy as np

sample_mean = 52
population_mean = 50
population_std = 5
sample_size = 30

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

alpha = 0.05
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H0): The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference between sample and population mean.")

```

```python
# Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
np.random.seed(42)
sample_size = 30
population_mean = 50
population_std = 5
alpha = 0.05

# Generate random sample data
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
sample_mean = np.mean(sample_data)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Critical values for two-tailed test
z_critical = stats.norm.ppf(1 - alpha / 2)

# Visualization
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")
plt.axvline(-z_critical, color='red', linestyle='dashed', label=f"Critical Value (-{z_critical:.2f})")
plt.axvline(z_critical, color='red', linestyle='dashed', label=f"Critical Value (+{z_critical:.2f})")
plt.axvline(z_score, color='green', linestyle='solid', label=f"Z-score ({z_score:.2f})")

plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3)
plt.fill_between(x, y, where=(x > -z_critical) & (x < z_critical), color='blue', alpha=0.3)

plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.show()

# Interpretation
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

if abs(z_score) > z_critical:
    print("Reject the null hypothesis (H0): The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference between sample and population mean.")
```

```python
#  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(alpha=0.05, effect_size=0.8, sample_size=30):
    population_mean = 50
    population_std = 5
    sample_std = population_std / np.sqrt(sample_size)

    x = np.linspace(45, 55, 1000)

    # Null hypothesis (H0) distribution
    h0_pdf = stats.norm.pdf(x, population_mean, sample_std)

    # Alternative hypothesis (H1) distribution (Shifted mean)
    h1_mean = population_mean + effect_size
    h1_pdf = stats.norm.pdf(x, h1_mean, sample_std)

    # Critical Z-value for alpha
    z_critical = stats.norm.ppf(1 - alpha)
    critical_value = population_mean + z_critical * sample_std

    plt.figure(figsize=(9, 5))

    plt.plot(x, h0_pdf, label="H0: Null Distribution", color="blue")
    plt.plot(x, h1_pdf, label="H1: Alternative Distribution", color="green")

    # Shade Type 1 Error (α)
    plt.fill_between(x, h0_pdf, where=(x >= critical_value), color='red', alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 Error (β)
    plt.fill_between(x, h1_pdf, where=(x <= critical_value), color='purple', alpha=0.3, label="Type 2 Error (β)")

    plt.axvline(critical_value, color='black', linestyle='dashed', label=f"Critical Value ({critical_value:.2f})")

    plt.legend()
    plt.title("Type 1 & Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.show()

# Run the function to visualize errors
visualize_errors()
```

```python
# Write a Python program to perform an independent T-test and interpret the results
import numpy as np
import scipy.stats as stats

# Generate two independent random samples
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)  # Sample 1 (mean=50, std=5)
group2 = np.random.normal(loc=53, scale=5, size=30)  # Sample 2 (mean=53, std=5)

# Perform independent T-test
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)

# Interpretation
alpha = 0.05  # Significance level

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H0): The two groups have significantly different means.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference between the groups.")
```

```python
# Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate paired sample data (before & after)
np.random.seed(42)
before = np.random.normal(loc=50, scale=5, size=30)  # Before treatment
after = before + np.random.normal(loc=2, scale=2, size=30)  # After treatment (slightly improved)

# Perform paired T-test
t_stat, p_value = stats.ttest_rel(before, after)

# Interpretation
alpha = 0.05  # Significance level
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis (H0): There is a significant difference before and after.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference before and after.")

# Visualization
plt.figure(figsize=(7, 5))
plt.boxplot([before, after], labels=['Before', 'After'], patch_artist=True,
            boxprops=dict(facecolor="skyblue"), medianprops=dict(color="red"))

plt.title("Paired Sample Comparison (Before vs. After)")
plt.ylabel("Values")
plt.show()
```

```python
# Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """Calculates the confidence interval for a sample mean."""
    sample_mean = np.mean(sample)  # Sample mean
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    n = len(sample)  # Sample size
    margin_error = stats.t.ppf((1 + confidence) / 2, df=n-1) * (sample_std / np.sqrt(n))  # Margin of error
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error
    return lower_bound, upper_bound

# Simulating a sample
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)  # Sample with mean=50, std=5, n=30

# Calculate and print the confidence interval
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
# Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    """Calculates the margin of error for a given confidence level."""
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased estimate)
    n = len(sample)  # Sample size
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # T-critical value
    moe = t_critical * (sample_std / np.sqrt(n))  # Margin of error
    return moe

# Simulating a sample
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=30)  # Sample with mean=50, std=5, n=30

# Calculate and print the margin of error
moe_value = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% Confidence Level): {moe_value:.4f}")
```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    """Calculates the posterior probability using Bayes' Theorem."""
    marginal_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / marginal_B
    return posterior_A_given_B

# Example Scenario: Disease Testing
prior_disease = 0.01  # P(A): 1% of the population has the disease
likelihood_positive_given_disease = 0.95  # P(B|A): Test correctly detects disease 95% of the time
prior_no_disease = 1 - prior_disease  # P(~A): 99% of the population does NOT have the disease
likelihood_positive_given_no_disease = 0.05  # P(B|~A): 5% false positive rate

# Compute Posterior Probability: P(Disease | Positive Test)
posterior = bayes_theorem(prior_disease, likelihood_positive_given_disease,
                          prior_no_disease, likelihood_positive_given_no_disease)

print(f"Probability of having the disease given a positive test result: {posterior:.4f}")
```

```python
#  Perform a Chi-square test for independence between two categorical variables in Python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Sample Data: Survey of Preferences by Gender
data = np.array([[50, 30],  # Males
                 [20, 40]]) # Females

# Create a Pandas DataFrame for better readability
df = pd.DataFrame(data, columns=['Likes Product A', 'Likes Product B'], index=['Male', 'Female'])

# Perform the Chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(df)

# Print Results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The two variables are dependent.")
else:
    print("Fail to reject the null hypothesis: The two variables are independent.")
```

```python
# Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution
import numpy as np
import scipy.stats as stats

# Observed frequencies from real data
observed = np.array([50, 30, 20])  # Example: Survey responses

# Expected frequencies (assuming equal probability distribution)
# Calculate expected frequencies to ensure they sum to the observed total
total_observed = np.sum(observed)
expected = np.array([total_observed / len(observed)] * len(observed))
# expected = np.array([33.33, 33.33, 33.33])  # If we expect uniform distribution

# Perform Chi-square Goodness-of-Fit Test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Print Results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The data does not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The data follows the expected distribution.")
```

```python
# Implement an F-test using Python to compare the variances of two random samples
import numpy as np
import scipy.stats as stats

# Generate random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10, Size=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std Dev=15, Size=30

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
F_statistic = var1 / var2 if var1 > var2 else var2 / var1  # Ensure F ≥ 1
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Compute p-value
p_value = 1 - stats.f.cdf(F_statistic, df1, df2)  # One-tailed test

# Print results
print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-Statistic: {F_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: Variances are equal.")
```

```python
# Write a Python program to perform an ANOVA test to compare means between multiple groups and
# interpret the result
import numpy as np
import scipy.stats as stats

# Generate random data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10, Size=30
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std Dev=10, Size=30
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std Dev=10, Size=30

# Perform one-way ANOVA test
F_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print results
print(f"F-Statistic: {F_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")
```

```python
# Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std Dev=10, Size=30
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std Dev=10, Size=30
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std Dev=10, Size=30

# Perform one-way ANOVA test
F_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Print results
print(f"F-Statistic: {F_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")

# Visualization: Box plot of the three groups
plt.figure(figsize=(8, 6))
sns.boxplot(data=[group1, group2, group3])
plt.xticks([0, 1, 2], ["Group 1", "Group 2", "Group 3"])
plt.title("Comparison of Group Means (One-Way ANOVA)")
plt.ylabel("Values")
plt.xlabel("Groups")
plt.grid(True)
plt.show()
```

```python
# Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import scipy.stats as stats

def check_anova_assumptions(*groups):
    """
    Check ANOVA assumptions: Normality and Homogeneity of Variance.

    :param groups: Lists of numerical data for different groups
    """

    # Normality Test (Shapiro-Wilk test)
    print("\n--- Normality Test (Shapiro-Wilk) ---")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: W-statistic={stat:.4f}, P-value={p_value:.4f}")
        if p_value < 0.05:
            print(f"  -> Group {i+1} is NOT normally distributed (p < 0.05).")
        else:
            print(f"  -> Group {i+1} is normally distributed (p ≥ 0.05).")

    # Homogeneity of Variance Test (Levene's test)
    print("\n--- Homogeneity of Variance Test (Levene’s Test) ---")
    stat, p_value = stats.levene(*groups)
    print(f"Levene’s Test: W-statistic={stat:.4f}, P-value={p_value:.4f}")
    if p_value < 0.05:
        print("  -> Variances are NOT equal (p < 0.05). Consider using Welch’s ANOVA.")
    else:
        print("  -> Variances are equal (p ≥ 0.05). Assumption holds.")

# Example: Generate random data and test assumptions
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
# Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
# results
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Simulate dataset
np.random.seed(42)
data = {
    "Factor_A": np.repeat(["Low", "Medium", "High"], 20),  # Factor A (e.g., Temperature)
    "Factor_B": np.tile(["Type 1", "Type 2"], 30),  # Factor B (e.g., Fertilizer Type)
    "Response": np.random.normal(50, 10, 60) + np.random.choice([0, 5, 10], 60)  # Dependent Variable (e.g., Plant Growth)
}

df = pd.DataFrame(data)

# Perform Two-Way ANOVA
model = ols("Response ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print ANOVA Results
print("\n--- Two-Way ANOVA Results ---")
print(anova_table)

# Visualization: Interaction Plot
plt.figure(figsize=(8, 6))
sns.pointplot(x="Factor_A", y="Response", hue="Factor_B", data=df, dodge=True, markers=["o", "s"], capsize=0.1)
plt.title("Two-Way ANOVA Interaction Plot")
plt.xlabel("Factor A (e.g., Temperature)")
plt.ylabel("Mean Response (e.g., Growth)")
plt.grid(True)
plt.show()
```

```python
# Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define degrees of freedom
df1, df2 = 5, 20  # df1 (numerator), df2 (denominator)
x = np.linspace(0, 5, 1000)  # Generate F-values

# Compute F-distribution PDF
y = stats.f.pdf(x, df1, df2)

# Plot F-distribution
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', label=f'F-distribution (df1={df1}, df2={df2})')
plt.fill_between(x, y, alpha=0.2, color='blue')
plt.axvline(stats.f.ppf(0.95, df1, df2), color='red', linestyle='dashed', label="Critical Value (α=0.05)")

# Labels and Title
plt.title("F-Distribution Curve")
plt.xlabel("F-Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
# Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating random data for 3 groups (Example: Test Scores of Students in 3 Classes)
np.random.seed(42)
group_A = np.random.normal(75, 10, 30)  # Mean=75, Std=10, n=30
group_B = np.random.normal(80, 10, 30)  # Mean=80, Std=10, n=30
group_C = np.random.normal(85, 10, 30)  # Mean=85, Std=10, n=30

# Combine data into a DataFrame
df = pd.DataFrame({
    "Scores": np.concatenate([group_A, group_B, group_C]),
    "Group": ["A"] * 30 + ["B"] * 30 + ["C"] * 30
})

# Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(group_A, group_B, group_C)

# Print ANOVA Results
print("\n--- One-Way ANOVA Results ---")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision based on p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject H₀: At least one group mean is significantly different.")
else:
    print("Fail to Reject H₀: No significant difference among group means.")

# Visualization: Boxplot for Group Comparison
plt.figure(figsize=(8, 5))
sns.boxplot(x="Group", y="Scores", data=df, palette="Set2")
plt.title("One-Way ANOVA: Group Comparison")
plt.xlabel("Groups")
plt.ylabel("Scores")
plt.grid()
plt.show()
```

```python
# Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
import scipy.stats as stats

# Simulate random data from a normal distribution
np.random.seed(42)
sample_size = 30
population_mean = 75  # Assumed population mean
population_std = 10   # Assumed population standard deviation
sample_data = np.random.normal(population_mean, population_std, sample_size)

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

# Perform hypothesis testing (T-test since sample size < 30)
t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)

# Print Results
print("\n--- Hypothesis Testing Results ---")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject H₀: The sample mean is significantly different from the population mean.")
else:
    print("Fail to Reject H₀: No significant difference between sample and population mean.")
```

```python
#  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

# Simulate random sample data
np.random.seed(42)
sample_size = 25
population_variance = 16  # Assumed population variance (σ²)
sample_data = np.random.normal(50, np.sqrt(population_variance), sample_size)

# Sample statistics
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (s²)

# Hypothesis Test (Chi-Square Test for Variance)
alpha = 0.05  # Significance level
df = sample_size - 1  # Degrees of freedom
chi_square_stat = (df * sample_variance) / population_variance

# Chi-square critical values
chi_critical_low = stats.chi2.ppf(alpha / 2, df)
chi_critical_high = stats.chi2.ppf(1 - alpha / 2, df)

# Print Results
print("\n--- Hypothesis Testing for Variance ---")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"Chi-Square Critical Range: ({chi_critical_low:.4f}, {chi_critical_high:.4f})")

# Decision Rule
if chi_square_stat < chi_critical_low or chi_square_stat > chi_critical_high:
    print("Reject H₀: The population variance is significantly different from the assumed variance.")
else:
    print("Fail to Reject H₀: No significant difference in variance.")
```

```python
# Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

# Sample Data: Number of successes and sample sizes for two groups
success_a = 60  # Group A (e.g., people who prefer product A)
n_a = 200       # Total sample size in Group A

success_b = 45  # Group B (e.g., people who prefer product B)
n_b = 180       # Total sample size in Group B

# Compute sample proportions
p_a = success_a / n_a
p_b = success_b / n_b

# Pooled proportion (combined proportion estimate)
p_pooled = (success_a + success_b) / (n_a + n_b)

# Standard error calculation
se = np.sqrt(p_pooled * (1 - p_pooled) * (1/n_a + 1/n_b))

# Compute Z-score
z_stat = (p_a - p_b) / se

# Compute p-value (Two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Print Results
print("\n--- Two-Proportion Z-Test ---")
print(f"Proportion in Group A: {p_a:.4f}")
print(f"Proportion in Group B: {p_b:.4f}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject H₀: The proportions are significantly different.")
else:
    print("Fail to Reject H₀: No significant difference between proportions.")
```

```python
# Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Generate two random datasets with different variances
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, StdDev=10
data2 = np.random.normal(loc=50, scale=15, size=100)  # Mean=50, StdDev=15

# Calculate sample variances
var1 = np.var(data1, ddof=1)  # Sample variance of dataset 1
var2 = np.var(data2, ddof=1)  # Sample variance of dataset 2

# Compute F-statistic and p-value
F_statistic = var1 / var2
df1 = len(data1) - 1  # Degrees of freedom for dataset 1
df2 = len(data2) - 1  # Degrees of freedom for dataset 2
p_value = 2 * (1 - stats.f.cdf(F_statistic, df1, df2))  # Two-tailed test

# Print Results
print("\n--- F-Test for Equality of Variances ---")
print(f"Variance of Data 1: {var1:.4f}")
print(f"Variance of Data 2: {var2:.4f}")
print(f"F-Statistic: {F_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject H₀: The variances are significantly different.")
else:
    print("Fail to Reject H₀: No significant difference in variances.")

# Visualization: Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data=[data1, data2])
plt.xticks([0, 1], ["Dataset 1", "Dataset 2"])
plt.title("Boxplot Comparison of Two Datasets")
plt.ylabel("Values")
plt.grid()
plt.show()
```

```python
#Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated observed data (e.g., counts from different categories)
observed = np.array([50, 42, 58, 60, 40])

# Expected frequencies (assuming equal distribution)
expected = np.full_like(observed, np.mean(observed))

# Perform Chi-Square Goodness-of-Fit Test
chi_square_stat, p_value = stats.chisquare(observed, expected)

# Print Results
print("\n--- Chi-Square Goodness-of-Fit Test ---")
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision Rule
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject H₀: The observed distribution is significantly different from the expected.")
else:
    print("Fail to Reject H₀: No significant difference between observed and expected.")

# Visualization: Bar Plot
categories = [f"Category {i+1}" for i in range(len(observed))]
x = np.arange(len(categories))

plt.figure(figsize=(8, 5))
plt.bar(x - 0.2, observed, width=0.4, label="Observed", color="blue")
plt.bar(x + 0.2, expected, width=0.4, label="Expected", color="orange", alpha=0.7)
plt.xticks(x, categories)
plt.xlabel("Categories")
plt.ylabel("Frequency")
plt.title("Chi-Square Goodness-of-Fit Test")
plt.legend()
plt.grid(axis="y")
plt.show()
```
