## Q1.What is NumPy, and why is it widely used in Python?

# Q2. How does broadcasting work in NumPy?
Broadcasting in NumPy is a feature that allows operations on arrays of different shapes without needing to manually replicate data. Here’s a concise overview:

How Broadcasting Works:
Shape Comparison: NumPy compares the shapes of the arrays from the trailing dimensions (rightmost) to the leading dimensions (leftmost).

Compatibility Rules:

*   Dimensions are compatible if they are equal.
*   One of the dimensions can be 1 (which can be stretched).

Stretching: If dimensions are not the same, the smaller array is "stretched" to match the larger array's shape without actually copying data.

Resulting Shape: The output shape is determined by taking the maximum size along each dimension.

#Q6. What are the differences between NumPy arrays and Python lists?
#1. Data Type Homogeneity
NumPy Arrays: All elements must be of the same data type (homogeneous).
Python Lists: Elements can be of different data types (heterogeneous).

# 2. Performance
 NumPy Arrays: Faster for numerical computations due to optimized performance.
 Python Lists: Slower for numerical operations, especially with large datasets.

# 3. Functionality
 NumPy Arrays: Provides a wide range of mathematical functions and supports broadcasting.
 Python Lists: Basic operations only; lacks built-in mathematical functions.

# 4. Memory Consumption
 NumPy Arrays: More memory-efficient due to contiguous memory allocation and fixed data types.
 Python Lists: Higher memory overhead due to flexibility in data types.

# 5. Multidimensional Support
 NumPy Arrays: Natively supports multi-dimensional arrays (e.g., 2D, 3D).
 Python Lists: Can simulate multi-dimensional arrays with nested lists, but less efficient.

# 6. Ease of Use
 NumPy Arrays: More complex syntax for advanced operations, designed for numerical computing.
 Python Lists: Simple and flexible, suitable for general-purpose programming.

#Q8. What does the term “vectorized operation” mean in NumPy?

# Q17. Explain the difference between apply() and map() in Pandas?
Scope: apply() can be used on both Series and DataFrames, while map() is primarily for Series.
Functionality: apply() can handle more complex functions that operate on rows or columns, while map() is for element-wise operations.
Output: apply() can return different shapes (Series or DataFrame), while map() always returns a Series of the same length as the input.

```python
#  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

numerical_iris = iris.select_dtypes(include=np.number)

correlation_matrix = numerical_iris.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Generate a bar plot using Plotly?

import plotly.graph_objects as go
import pandas as pd


data = {
    'Fruits': ['Apples', 'Bananas', 'Cherries', 'Dates'],
    'Quantities': [10, 15, 7, 5]
}

df = pd.DataFrame(data)

fig = go.Figure(data=[
    go.Bar(x=df['Fruits'], y=df['Quantities'], marker_color='indianred')
])


fig.update_layout(
    title='Fruit Quantities',
    xaxis_title='Fruits',
    yaxis_title='Quantities',
    template='plotly_white'
)

# Step 5: Show the plot
fig.show()
```

```python
# Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)


y1 = 2 * x + 1
y2 = x**2

plt.figure(figsize=(8, 5))

plt.plot(x, y1, label='Linear: y = 2x + 1', color='blue')
plt.plot(x, y2, label='Quadratic: y = x^2', color='orange')

plt.title('Simple Line Plot with Multiple Lines')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()


plt.show()


plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 22],
    'Salary': [50000, 60000, 70000, 80000, 45000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

threshold = 60000
filtered_df = df[df['Salary'] > threshold]

print("\nFiltered DataFrame (Salary > 60000):")
print(filtered_df)
```

```python
# Perform matrix multiplication using NumPy
import numpy as np

# Matrix A (2x3)
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

result_dot = np.dot(A, B)

result_at = A @ B

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nResult of A dot B (using np.dot()):")
print(result_dot)
print("\nResult of A dot B (using @ operator):")
print(result_at)
```

```python
# Create a 3D scatter plot using Plotly
import plotly.graph_objects as go
import numpy as np

np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z,
    mode='markers',
    marker=dict(size=5, color=z, colorscale='Viridis', opacity=0.8)
)])

fig.update_layout(title='3D Scatter Plot', scene=dict(
    xaxis_title='X Axis',
    yaxis_title='Y Axis',
    zaxis_title='Z Axis'
))

fig.show()
```
