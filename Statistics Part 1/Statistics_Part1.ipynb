# What is a random variable in probability theory?
A random variable is a function that assigns a numerical value to each possible outcome of a random experiment.

## What are the types of random variables?



*  Discrete Random Variable: Takes countable values (e.g., number of heads in coin tosses).
* Continuous Random Variable: Takes uncountable values within a range (e.g., height of students).  



# What is the difference between discrete and continuous distributions?


*   Discrete Distribution: Deals with countable values (e.g., number of students in a class).

*   Continuous Distribution: Deals with infinite values in a range (e.g., temperature measurements).



# What are probability distribution functions (PDFs)?
A Probability Distribution Function (PDF) defines the likelihood of a random variable taking a particular value.


*   For Discrete Variables: Called the Probability Mass Function (PMF).

*   For Continuous Variables: Called the Probability Density Function (PDF).




# How do cumulative distribution functions (CDFs) differ from probability distribution functions (PDFs)?

*   PDF: Shows the probability of a random variable at a specific point.

*  CDF: Gives the probability that a random variable is less than or equal to a given  value.

#What is a discrete uniform distribution?
 A discrete uniform distribution assigns equal probability to all possible outcomes.

Example: Rolling a fair 6-sided die (each number has a probability of 1/6).

# What are the key properties of a Bernoulli distribution?








*  It has two possible outcomes: Success (1) or Failure (0).
*  Probability of success = p, probability of failure = 1 - p.
*Mean = p, Variance = p(1-p).

Example: Flipping a fair coin (Heads = 1, Tails = 0).

# What is the binomial distribution, and how is it used in probability?
The binomial distribution is a discrete probability distribution that models the probability of a certain number of successes in a fixed number of independent trials, each with the same probability of success

## What is the Poisson distribution, and where is it applied?
The Poisson distribution is a discrete probability distribution that models the probability of a certain number of events occurring within a fixed interval of time or space, given a known average rate of occurrence. It's commonly applied in fields like queueing theory, epidemiology, and telecommunications

# What is a continuous uniform distribution?
A continuous uniform distribution, also known as a rectangular distribution, is a probability distribution where all values within a specified range are equally likely to occur, meaning the probability density function (PDF) is constant over that range.

# What are the characteristics of a normal distribution?

* Bell-shaped curve, symmetric about the mean (μ).
* Mean = Median = Mode.
* 68%-95%-99.7% Rule:
* 68% of data lies within 1 standard deviation (σ).
* 95% within 2σ, 99.7% within 3σ.

Example: Heights of people, IQ scores, blood pressure levels.

# What is the standard normal distribution, and why is it important?

* A normal distribution with mean = 0 and standard deviation = 1.
* Used to compute Z-scores, making it easy to compare different datasets.
* Helps in hypothesis testing and probability calculations.

# What is the Central Limit Theorem (CLT), and why is it critical in statistics?

* CLT states: The sampling distribution of the mean approaches a normal distribution as the sample size increases, regardless of the population's shape.
* Importance:
1.   Allows us to make inferences about a population from sample data.
2.   Forms the foundation for confidence intervals and hypothesis testing.



# How does the Central Limit Theorem relate to the normal distribution?

* The CLT ensures that the mean of a sufficiently large sample follows a normal distribution, even if the population data is not normally distributed.
* The larger the sample size (n ≥ 30), the more the sample mean approximates a normal curve.

# What is the application of Z statistics in hypothesis testing?

* Used in Z-tests to compare a sample mean to a population mean when:
* The population variance is known.
* The sample size is large (n ≥ 30).

Example: Testing whether the average height of students differs from the national average.

# What are point estimates and interval estimates in statistics?

* Point Estimate: A single value used to estimate a population parameter (e.g., sample mean x̄).
* Interval Estimate: A range of values (e.g., confidence interval) that likely contains the population parameter.

# What is the significance of confidence intervals in statistical analysis?

* A confidence interval (CI) gives a range of values within which the true population parameter is likely to fall.

Example:
A 95% CI for the mean weight of students (60 ± 2 kg) means we are 95% confident the true mean is between 58 kg and 62 kg.
* Higher confidence level → wider interval.

# What is the relationship between a Z-score and a confidence interval?
A Z-score determines how many standard deviations a data point is from the mean, while a confidence interval uses this Z-score, along with other data, to create a range where the population parameter is likely to fall, with a specified level of confidence.

# How are Z-scores used to compare different distributions?

* Z-scores standardize values by converting different distributions into a common scale (mean = 0, standard deviation = 1).
* Allows comparison of scores from different datasets with different means and standard deviations.

Example: Comparing SAT scores (mean = 1050, σ = 200) with IQ scores (mean = 100, σ = 15).

# What are the assumptions for applying the Central Limit Theorem (CLT)?

* Random Sampling: Data must be randomly selected.
* Independence: Each observation should be independent of others.
* Sample Size: Should be large (n ≥ 30) for the theorem to hold well.
* Finite Variance: Population should have a finite standard deviation.

# What is the concept of expected value in a probability distribution?
In a probability distribution, the expected value (also known as the mean or average) is the weighted average of all possible outcomes, where each outcome is weighted by its probability.

# How does a probability distribution relate to the expected outcome of a random variable?
A probability distribution describes the likelihood of different outcomes for a random variable, and the expected outcome is calculated by weighting each possible value by its probability and summing these products.

```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    return bernoulli.pmf(x, p)

# Example: Bernoulli Distribution with p = 0.5
p = 0.5
x_values = [0, 1]
pdf_values = [bernoulli_pdf(p, x) for x in x_values]

print("Bernoulli PDF:", pdf_values)
```

```python
# Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

def standard_normal_distribution():
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)

    plt.plot(x, y, label="Standard Normal PDF")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Standard Normal Distribution (Mean=0, Std=1)")
    plt.legend()
    plt.grid()
    plt.show()

standard_normal_distribution()

```

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
# distribution
import numpy as np
import scipy.stats as stats

def calculate_z_score(value, mean, std_dev):
    return (value - mean) / std_dev

# Example dataset
data = np.random.normal(loc=50, scale=10, size=100)
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-score for a given value
value = 60
z_score = calculate_z_score(value, mean, std_dev)

print(f"Z-score of {value}: {z_score}")
print(f"Probability from Standard Normal Distribution: {stats.norm.cdf(z_score)}")
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats

# Sample data
sample = np.random.normal(loc=52, scale=10, size=30)
population_mean = 50
population_std_dev = 10
sample_mean = np.mean(sample)
sample_size = len(sample)

# Calculate Z-score
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# P-value for hypothesis test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Z-Statistic: {z_statistic}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#  Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

data = np.random.normal(loc=50, scale=10, size=100)
confidence = 0.95
mean = np.mean(data)
std_err = stats.sem(data)
margin_of_error = stats.t.ppf((1 + confidence) / 2., len(data)-1) * std_err

ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats

data = np.random.normal(loc=100, scale=15, size=200)

mean = np.mean(data)
std_err = stats.sem(data)
ci = stats.norm.interval(0.95, loc=mean, scale=std_err)

print(f"95% Confidence Interval for Mean: {ci}")
```

```python
# Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate two datasets
data1 = np.random.normal(loc=50, scale=5, size=1000)
data2 = np.random.normal(loc=50, scale=15, size=1000)

# Compute standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Visualization
plt.figure(figsize=(10, 6))
sns.kdeplot(data1, label="Dataset 1 (Lower Std Dev)", linewidth=2)
sns.kdeplot(data2, label="Dataset 2 (Higher Std Dev)", linewidth=2)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Comparison of Two Datasets with Different Standard Deviations")
plt.legend()
plt.show()
```

```python
# Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np
import scipy.stats as stats

# Generate a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate interquartile range (IQR)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
# Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
# distribution.
import numpy as np
import scipy.stats as stats

def compute_skewness_kurtosis():
    # Generate normal distribution dataset
    data = np.random.normal(loc=50, scale=10, size=1000)

    # Calculate skewness and kurtosis
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True returns excess kurtosis

    print(f"Skewness: {skewness:.2f}")
    print(f"Kurtosis: {kurtosis:.2f}")

compute_skewness_kurtosis()
```
